```python
r = requests.get(url)

status = r.status_code
encoding = r.encoding
html_doc = r.text

soup = BeautifulSoup(html_doc, 'html.parser')
anchor = soup(html_target)
```

```python
#print(soup.prettify())
right_td=soup.find('td',class_="SousCadre")
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]


for row in right_td.findAll("tr",class_="LigneMatch"):
    cells = row.findAll("td")
    A.append(str(cells[1].text))
    B.append(str(cells[3].text))
    C.append(str(cells[4].text))
    D.append(str(cells[5].text))
    E.append(str(cells[7].text))
    F.append(str(cells[9].text))
    G.append(str(cells[10].text))
    H.append(str(cells[12].text))
 
```

```python
df=pd.DataFrame(A,columns=['Journee'])
df['Domicile']=B
df['Résultat']=C
df['exterieur']=D
df['Dom_place_journee']=E
df['Ext_place_journee']=F
df['Dom_place_finale']=G
df['Ext_place_finale']=H


file_name="Classement.csv"

#df.to_csv(file_name)
pd.read_csv(file_name).append(df).drop_duplicates().to_csv(file_name, sep=',', encoding='utf-8',index=False)

df
```

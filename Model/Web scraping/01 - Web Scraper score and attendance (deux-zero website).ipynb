```python
import requests
from bs4 import BeautifulSoup
import os
#import pandas to convert list to data frame
import pandas as pd
```

```python
html_target = "a"
tag = "href"
f_ext = ".pdf"
dir_name = "Ghodsi_Ali"
url = "http://www.deux-zero.com/index.php?Code=L1&Item=Stades&Edition=2017-2018&StadeId=229"
```

```python
r = requests.get(url)

status = r.status_code
encoding = r.encoding
html_doc = r.text

soup = BeautifulSoup(html_doc, 'html.parser')
anchor = soup(html_target)
```

```python
#print(soup.prettify())
right_td=soup.find('td',class_="SousCadre")
```

```python
G=[]
tbody = soup.findAll(class_='Txt11')[1]
tbody
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]


for row in right_td.findAll("tr",class_="LigneMatch"):
    cells = row.findAll("td")
    A.append(str(cells[0].text))
    B.append(str(cells[2].text))
    C.append(str(cells[3].text))
    D.append(str(cells[4].text))
    E.append(str(cells[6].text))
    F.append(str(cells[6].text))
    G.append(tbody.text)
 
```

```python
df=pd.DataFrame(A,columns=['Date'])
df['Domicile']=B
df['Résultat']=C
df['exterieur']=D
df['Affluence']=E
df['Capacite']=G

df['Date'].apply(lambda x: x[len(x)+0:len(x)-5])

df['Capacite'] = df['Capacite'].str.replace('places','')
df['Capacite'] = df['Capacite'].str[:5]

file_name="LYON.csv"

#df.to_csv(file_name)
pd.read_csv(file_name).append(df).drop_duplicates().to_csv(file_name, sep=',', encoding='utf-8',index=False)

df
```
